### Data Cleaning

>**Dropping duplicates and unwanted row**

```python
#dropping duplicates
db.drop_duplicates(inplace=True) 
db.duplicated().sum() # confirming that there is no duplicate.
```

```python
# remove unwanted row
db = db.drop(Unwanted_col, axis =1)
db.head(2) # previewing
```

>**Dropping rows with zeros (0) in revenue and budget column**

```python
#dropping zero row in revenue and budget column
db_rev_0 = db.query('revenue == 0') # creating dataframe for revenue with with zero 
db_bud_0 = db.query('budget == 0') # creating dataframe for budget with zero value.
db_rev_0.shape, db_bud_0.shape #showing the initial shape for later verification
```

```python
col_0 = ['revenue','budget'] #creating a list of column
db[col_0] = db[col_0].replace(0,np.NaN) #replacing zero with NaN using np.NaN
db.dropna(subset=col_0, inplace=True) # dropping the replaced NaN from the dataframe.
```

```python
# rechecking for confirmation
db_bud_0 = db.query('budget == 0') 
db_rev_0 = db.query('revenue == 0')
db_rev_0.shape, db_bud_0.shape 
```

>**Checking Datatype for Revenue, Budget and Release date**

```python
# further investigation on the revenue and budget datatype before creating profit column
type(db.revenue[0]),type(db.budget[0]) 
```

```python
# converting both column to "int" datatype
db.revenue = db.revenue.astype(int) 
db.budget = db.budget.astype(int)

db.revenue[0].dtype, db.budget[0].dtype #checking for confrimation
```

```python
# changing release date data type to date format
db.release_date = pd.to_datetime(db.release_date)

db.head(2) #previewing 
```

>**Creating Profit and Return of Interest (ROI) column**

```python
# Creating the profit column
db.insert(3, 'profit', db.revenue - db.budget) 
# Creating the return of interest column(ROI) column
db.insert(4, 'roi', db.profit / db.budget)
db.head(2) # Previewing
```

>**Creating a function that split a column that has pipe ('|')**

```python
# a function to split column having a pipe (|) as a sepearator
def split(col):
    details = db[col].str.cat(sep='|')
    content = pd.Series(details.split('|'))
    
    return content

a = split('director')
a.value_counts()
```

<a id='ra7'></a>
>**Which is the most popular genre**
>
>To get the most popular genre, we'll be going through some steps as seen below to achiecve this.

```python
dbv2 = db.copy() #making a copy of the main data set for further analysis on popularity
dbv2.head(2) #previewing data
```

>**Step 1: Turn Genres rows to list**

```python
split = ['genres'] # assigning genre to a variable
#function to split genre column
for c in split:
    dbv2[c] = dbv2[c].apply(lambda x: x.split("|"))
    
dbv2.head(3) #previewing data
```

>**Step 2: Make a Crosstab of popularity with genres**

```python
#reproducing the content of the genre column as a row, thus duplicating the index

genre_cross = dbv2.explode('genres') #exploding genre column
uni_genre=genre_cross.genres.unique() # creating a list of unique genre for later iteration

#creating a crosstab to show frquency of genre with the group of population
genre_cross = pd.crosstab(genre_cross['popularity'],genre_cross['genres']).reset_index().rename_axis(columns=None)

print(uni_genre) #previewing list of unique genre
genre_cross.head(3) #previewing new crosstab table

```

>**Step 3: Calculate the popularity of each genres by creating a function that iterates over the unique element of the genre column. Then form an array of genre**

```python
#function to calcalute the sum of popularity for each genre
genre_names = [] #empty list for appending genre name
genre_pop = [] #empty list for appending genre population
for genre in uni_genre: #iterating over list of genre
    pop_gen = genre_cross.loc[:,['popularity',genre]] #slicing the dataframe to select popularity and a genre
    pop_gen[genre] = pop_gen[genre].replace(0,np.NaN) #replacing zeros(0) in columns of genre with numpy NaN 
    pop_gen.dropna(subset=genre, inplace=True) #dropping rows with null values (NaN)
    y = pop_gen.popularity.sum() #summing the popularity after the rows with zero are dropped
    genre_names.append(genre) #filling the list with each genre 
    genre_pop.append(y) #filling the list with the summed popularity
    
print(genre_names,genre_pop) #previewing our two new list

# joining both list together with genres on the left side and population to the right
genre_pop_array =np.column_stack((genre_names,genre_pop)) 
genre_pop_array
```

>**Step 4: Converting the array to a dataframe, change the datatype and sort the popularity column**

```python
#converting the list to a dataframe for further analysis
gen_pop_df = pd.DataFrame(genre_pop_array,columns=['genres','popularity'])
gen_pop_df.head(3)
```

```python
#converting the datatype to float first because of the initial error
gen_pop_df['popularity'] = gen_pop_df['popularity'].astype(float)

#changing from the former datatype(float) to datatype(int)
gen_pop_df['popularity'] = gen_pop_df['popularity'].astype(int)

gen_pop_df.dtypes #confirming assigned datatype
```

```python
#sorting new dataframe in ascending order
gen_pop_df.sort_values(by='popularity',inplace=True)

#reseting the index to start from 0
gen_pop_df.reset_index(inplace=True,drop=True)
gen_pop_df.head(3)
```

>**Step 5: Visualizing the dataframe**

```python
#Visualizing the dataframe to show the most popular genre
#plotting a horizontal bar chart
gp = gen_pop_df.plot.barh(x='genres',y='popularity',color='Blue',figsize=(12,6))
gp.set_title('Genres Popularity', color='DarkRed', fontsize=17) #naming title
gp.set_xlabel('Popularity', color='DarkRed',fontsize=15) #naming horizotal axis
gp.set_ylabel('Genres', color='DarkRed',fontsize=15) #naming vertical axis
gp.tick_params(labelsize=12, labelcolor='DarkRed'); #setting both tick parameters
```

### 3. Monetary Analysis

>**Creating a function that calculate minimum and maximum value for a column and concatenate both and return a dataframe.**

```python
def check(col): #defining function name
    min_id = db[col].idxmin() #selecting the id of the minimum value 
    min_df = pd.DataFrame(db.loc[min_id]) # selecting the row with minimum value and saving it as a dataframe 
    max_id = db[col].idxmax() #selecting the id of the maximum value 
    max_df = pd.DataFrame(db.loc[max_id]) #selecting the row with minimum valu and saving it as a dataframe
    min_max = pd.concat([min_df,max_df], axis =1) #joining both dataframe
    
    return min_max #returning the joined dataframe when the function is called
```

>**7. Top Budget Movie**

```python
mbr = db.loc[:,['original_title','budget','revenue']] #selecting movie name,budget and revenue column
mbr.head(3) #previewing
```

```python
#function that returns a dataframe sorted by the given column
def slice(col): # defining function
    mbr.sort_values(by=col,ascending=False,inplace=True) # sort dataframe by specific column in descending order
    mbr.reset_index(inplace=True,drop=True) #rearrange the index after sorting
    ten_row = mbr.iloc[0:10,:] #select the first ten row of the column
    srtd_ten_row = ten_row.copy() #make a copy for sorting
    srtd_ten_row.sort_values(by=col,inplace=True) #sort the column in ascending order for plotting
    srtd_ten_row.reset_index(inplace=True,drop=True) #reset the index for good arrangement in disaplay
    return srtd_ten_row #return a sorted dataframe when the function is called
```

```python
# calling the function to create the top 10 movie with high budget
mov_bud = slice('budget')  

#Visualizing the dataframe to show the most budget movie
#plotting a horizontal bar chart
mb = mov_bud.plot.barh(x='original_title',y='budget',color='Blue',figsize=(10,5))
mb.set_title('Movie Budgets', color='DarkRed', fontsize=17) #naming title
mb.set_xlabel('Budget [in Hundred Million $]', color='DarkRed',fontsize=15) #naming horizotal axis
mb.set_ylabel('') #naming vertical axis
mb.tick_params(labelsize=12, labelcolor='DarkRed'); #setting both tick parameters
```
